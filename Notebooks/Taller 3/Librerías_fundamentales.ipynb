### **¿Qué es NumPy?** 

NumPy **(Numerical Python)** un paquete fundamental para la computación científica en Python. Nos provee un objeto de matriz multidimensional (**multidimensional array object**) y una variedad de rutinas para operaciones rápidas sobre arreglos, que incluyen matemáticas, lógica, manipulación de ***shapes*** (tupla que le da una indicación del número de dimensiones en la matriz.), clasificación, selección, E / S, transformadas discretas de Fourier, álgebra lineal básica, operaciones estadísticas básicas, simulación aleatoria y mucho más. $^1$

La potencia del paquete de Numpy está basado en el objeto ***ndarray***. Este encapsula arreglos n-dimensionales de tipos de datos homogéneos, con muchas operaciones que se realizan con **vectorización, código C y Fortran pre-compilado** para el rendimiento. Hay varias diferencias importantes entre las matrices NumPy y las secuencias estándar de Python:

 Hay varias diferencias importantes entre las matrices NumPy y las secuencias estándar de Python:

1.   el código vectorizado es más conciso y más fácil de leer

2.   menos líneas de código generalmente significan menos errores

3.   el código se parece más a la notación matemática estándar (lo que facilita, por lo general, codificar correctamente las construcciones matemáticas)

#### Suma y multplicación de matrices con NumPy

````
X = [[12,7,3],
    [4 ,5,6],
    [7 ,8,9]]

Y = [[5,8,1],
    [6,7,3],
    [4,5,9]]
````

````
import time

result = [[0,0,0],
         [0,0,0],
         [0,0,0]]

start_time = time.time()

# iterate en las filas
for i in range(len(X)):
   # iterate en las columnas
   for j in range(len(X[0])):
       result[i][j] = X[i][j] + Y[i][j]

for r in result:
   print(r)
print("--- %s segundos ---" % (time.time() - start_time))
````

### **Vectorización**

La vectorización es una habilidad poderosa dentro de NumPy para expresar operaciones como si ocurrieran en arreglos completos en lugar de sus elementos individuales. Una definición concisa:

"Esta práctica de reemplazar bucles explícitos con expresiones de matriz se conoce comúnmente como vectorización. En general, las operaciones de matriz vectorizada a menudo serán uno o dos (o más) órdenes de magnitud más rápidas que sus equivalentes de Python puro, con el mayor impacto -visto- en cualquier tipo de cálculo numérico. [[fuente](https://www.oreilly.com/library/view/python-for-data/9781449323592/ch04.html)]"

Al recorrer una matriz o cualquier estructura de datos en Python, hay mucha sobrecarga involucrada. Las operaciones vectorizadas en NumPy delegan el bucle internamente a funciones C y Fortran altamente optimizadas, lo que hace que el código Python sea más limpio y rápido.

**Código vectorizado vs código no vectorizado**

$$ Z = W^{T} X $$

````
import matplotlib.pyplot as plt
import numpy as np
#Constants
h = 6.626e-34
m = 9.11e-31
#Values for L and x
x_list = np.linspace(0,1,100)
L = 1

def psi(n,L,x):
    return np.sqrt(2/L)*np.sin(n*np.pi*x/L)

def psi_2(n,L,x):
    return np.square(psi(n,L,x))
    
plt.figure(figsize=(15,10))
plt.suptitle("Wave Functions", fontsize=18)
for n in range(1,4):
    #Empty lists for energy and psi wave
    psi_2_list = []
    psi_list = []
    for x in x_list:
        psi_2_list.append(psi_2(n,L,x))
        psi_list.append(psi(n,L,x))
    plt.subplot(3,2,2*n-1)
    plt.plot(x_list, psi_list)
    plt.xlabel("L", fontsize=13)
    plt.ylabel("Ψ", fontsize=13)
    plt.xticks(np.arange(0, 1, step=0.5))
    plt.title("n="+str(n), fontsize=16)
    plt.grid()
    plt.subplot(3,2,2*n)
    plt.plot(x_list, psi_2_list)
    plt.xlabel("L", fontsize=13)
    plt.ylabel("Ψ*Ψ", fontsize=13)
    plt.xticks(np.arange(0, 1, step=0.5))
    plt.title("n="+str(n), fontsize=16)
    plt.grid()
plt.tight_layout(rect=[0, 0.03, 1, 0.95])
````
